$$s(t) = a1 + a2 \cdot arctan(x − 1) + a3 \cdot (x + 1)^{3}$$

<center><h4>2. Aplicar el método de Newton-Raphson para determinar la raíz (es decir, el cruce por cero) de la señal.  </h4></center>
<center><h4>3. Estimar la primera derivada de la señal empleando el método de diferencias finitas hacia adelante.  </h4></center>
<center><h4>4. Calcular la integral de la señal en el rango [0, 5] utilizando el método de integración trapezoidal.  </h4></center>
